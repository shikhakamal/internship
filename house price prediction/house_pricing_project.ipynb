TRAINING PROCESS

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
x
```

```python
y
```

cheking for max random state

```python
maxAcc=0
maxRS=0

for i in range(1,300):
    x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    y_pred=lr.predict(x_test)
    score=r2_score(y_test, y_pred)
    print('score is ',score,' random state is ',i)
    
    if score>maxAcc:
        maxAcc=score
        maxRS=i
        print('max score is ', maxAcc,' at random state ',maxRS)
```

max score is 89.42 % and max random state is 266

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=266)
```

```python
lr=LinearRegression()
dtr=DecisionTreeRegressor()
rfr=RandomForestRegressor()
knn=KNeighborsRegressor()
svm=SVR()

list=[lr, dtr, rfr, knn, svm]
for i in list:
    i.fit(x_train, y_train)
    y_pred=i.predict(x_test)
    score=r2_score(y_test, y_pred)
    
    print('\n\n\n The score of ',i,'is ', score)
    
```

CROSS VALIDATION SCORE

```python
from sklearn.model_selection import cross_val_score
```

```python
for i in list:
    scr=cross_val_score(i, x, y, cv=7)
    print('\n\n\n The cross val score of ', i,'is ',scr.mean())
```

the random forest model is best fit in above all rest models, so we continue with random forest model only
